```python
def load_activity_map():
    map = {}
    map[0] = 'transient'
    map[1] = 'lying'
    map[2] = 'sitting'
    map[3] = 'standing'
    map[4] = 'walking'
    map[5] = 'running'
    map[6] = 'cycling'
    map[7] = 'Nordic_walking'
    map[9] = 'watching_TV'
    map[10] = 'computer_work'
    map[11] = 'car driving'
    map[12] = 'ascending_stairs'
    map[13] = 'descending_stairs'
    map[16] = 'vacuum_cleaning'
    map[17] = 'ironing'
    map[18] = 'folding_laundry'
    map[19] = 'house_cleaning'
    map[20] = 'playing_soccer'
    map[24] = 'rope_jumping'
    return map
```

```python
def generate_three_IMU(name):
    x = name +'_x'
    y = name +'_y'
    z = name +'_z'
    return [x,y,z]

def generate_four_IMU(name):
    x = name +'_x'
    y = name +'_y'
    z = name +'_z'
    w = name +'_w'
    return [x,y,z,w]

def generate_cols_IMU(name):
    # temp
    temp = name+'_temperature'
    output = [temp]
    # acceleration 16
    acceleration16 = name+'_3D_acceleration_16'
    acceleration16 = generate_three_IMU(acceleration16)
    output.extend(acceleration16)
    # acceleration 6
    acceleration6 = name+'_3D_acceleration_6'
    acceleration6 = generate_three_IMU(acceleration6)
    output.extend(acceleration6)
    # gyroscope
    gyroscope = name+'_3D_gyroscope'
    gyroscope = generate_three_IMU(gyroscope)
    output.extend(gyroscope)
    # magnometer
    magnometer = name+'_3D_magnetometer'
    magnometer = generate_three_IMU(magnometer)
    output.extend(magnometer)
    # oreintation
    oreintation = name+'_4D_orientation'
    oreintation = generate_four_IMU(oreintation)
    output.extend(oreintation)
    return output

def load_IMU():
    output = ['time_stamp','activity_id', 'heart_rate']
    hand = 'hand'
    hand = generate_cols_IMU(hand)
    output.extend(hand)
    chest = 'chest'
    chest = generate_cols_IMU(chest)
    output.extend(chest)
    ankle = 'ankle'
    ankle = generate_cols_IMU(ankle)
    output.extend(ankle)
    return output
```

```python

def load_subjects(root='subject'):
    output = pd.DataFrame()
    cols = load_IMU()
    
    dataframes = []  # List to store individual dataframes
    for i in range(101, 110):
        path = root + str(i) + '.dat'
        subject = pd.read_table(path, header=None, sep='\s+')
        subject.columns = cols
        subject['id'] = i
        dataframes.append(subject)  # Append the dataframe to the list
    
    output = pd.concat(dataframes, ignore_index=True)  # Concatenate all dataframes
    output.reset_index(drop=True, inplace=True)
    return output

data = load_subjects()
```

```python
activity_map = {
    0: 'transient',
    1: 'lying',
    2: 'sitting',
    3: 'standing',
    4: 'walking',
    5: 'running',
    6: 'cycling',
    7: 'Nordic_walking',
    9: 'watching_TV',
    10: 'computer_work',
    11: 'car driving',
    12: 'ascending_stairs',
    13: 'descending_stairs',
    16: 'vacuum_cleaning',
    17: 'ironing',
    18: 'folding_laundry',
    19: 'house_cleaning',
    20: 'playing_soccer',
    24: 'rope_jumping'
}
df_new['activity_id'] = df_new['activity_id'].map(activity_map)

print(df_new[['activity_id']].head(10))  # Print the first 10 rows of the 'activity_id' column
```

```python
import numpy as np

# Define activities to remove
activities_to_drop = ['playing_soccer', 'computer_work', 'car driving', 'running', 'transient']

# Filter out unwanted activities and NaN values
df_filtered = df_new[~df_new['activity_id'].isin(activities_to_drop)]  # Drop specified activities
df_filtered = df_filtered.dropna(subset=['activity_id'])  # Remove NaN values

# Print unique activities in the filtered dataset
print("Unique activities in the filtered dataset:", df_filtered['activity_id'].unique())
```

```python
# Define activity replacements
activity_replacements = {
    'vacuum_cleaning': 'standing_household_chores',
    'ironing': 'standing_household_chores',
    'folding_laundry': 'standing_household_chores',
    'house_cleaning': 'standing_household_chores',
    'watching_TV': 'sitting',
    'rope_jumping': 'jumping'  # Rename 'rope_jumping' to 'jumping'
}

# Apply replacements
df_filtered['activity_id'] = df_filtered['activity_id'].replace(activity_replacements)

# Print unique activities to verify changes
print("Unique activities after merging & renaming:", df_filtered['activity_id'].unique())
```

```python
unique_activities = df_filtered['activity_id'].unique()

# Print all unique activity IDs
print(unique_activities)
```

```python
# Rename the column 'activity_id' to 'label'
df_filtered = df_filtered.rename(columns={'activity_id': 'label'})

# Print unique labels to verify
unique_activities = df_filtered['label'].unique()
print("Unique activities after renaming:", unique_activities)
```
