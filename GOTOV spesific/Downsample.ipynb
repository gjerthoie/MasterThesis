```python
# Define full paths to the files
train_file = os.path.join(directory_path, 'train_data.csv')
test_file = os.path.join(directory_path, 'test_data.csv')

# Load the CSV files
train_data = pd.read_csv(train_file)
test_data = pd.read_csv(test_file)

# Print to verify
print("Train shape:", train_data.shape)
print("Test shape:", test_data.shape)
```

```python
import pandas as pd

# Adjust label count for train, test, and train_2
train_limited = train_data.groupby('label', group_keys=False).head(10000)
test_limited = test_data.groupby('label', group_keys=False).head(5000)

def upsample_min_labels(df, min_samples=5000):
    balanced = []
    for label, group in df.groupby('label'):
        if len(group) < min_samples:
            # Repeat rows to reach minimum
            reps = (min_samples + len(group) - 1) // len(group)
            group_upsampled = pd.concat([group] * reps, ignore_index=True).head(min_samples)
        else:
            group_upsampled = group
        balanced.append(group_upsampled)
    return pd.concat(balanced, ignore_index=True)

# Apply to train and test datasets
train_limited = upsample_min_labels(train_limited, min_samples=5000)
test_limited = upsample_min_labels(test_limited, min_samples=5000)


# Print new label distributions
print("New label counts in train:\n", train_limited['label'].value_counts())
print("\nNew label counts in test:\n", test_limited['label'].value_counts())


```

```python
train = train_limited
```

```python
test = test_limited
```

```python
# get X_train and y_train from csv files
X_train = train.drop(['user_id', 'label'], axis=1)
y_train = train.label
```

```python

# get X_test and y_test from test csv file
X_test = train.drop(['user_id', 'label'], axis=1)
y_test = test.label
```

```python
# Separate features and labels for training data
X_train = train.drop(columns=['label'])  # Drop the label column to get features
y_train = train['label']  # Select only the label column for targets

# Separate features and labels for testing data
X_test = test.drop(columns=['label'])  # Drop the label column to get features
y_test = test['label']  # Select only the label column for targets

# Print shapes to confirm
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
label = train['label'].unique().tolist()
```

## LR

```python


# Set up Grid Search for Logistic Regression
parameters = {'C': [0.01, 0.1, 1, 10, 20, 30], 'penalty': ['l2', 'l1']}
log_reg = linear_model.LogisticRegression(solver='liblinear')
log_reg_grid = GridSearchCV(log_reg, param_grid=parameters, cv=3, verbose=1, n_jobs=-1)

log_reg_grid_results = perform_model(log_reg_grid, X_train, y_train, X_test, y_test, class_labels=label)
print_grid_search_attributes(log_reg_grid)
```

```python

# and labels is defined with the names of the classes
plt.figure(figsize=(8, 8))
plt.grid(False)
plot_confusion_matrix(log_reg_grid_results['confusion_matrix'], classes=label, normalize=False, title='Confusion matrix', cmap=plt.cm.Greens)
plt.show()
```

## RF


```python
from sklearn.ensemble import RandomForestClassifier
params = {'n_estimators': np.arange(10,201,20), 'max_depth':np.arange(3,15,2)}
rfc = RandomForestClassifier()
rfc_grid = GridSearchCV(rfc, param_grid=params, n_jobs=-1)
rfc_grid_results = perform_model(rfc_grid, X_train, y_train, X_test, y_test, class_labels=label)
```

```python
plt.figure(figsize=(8, 8))
plt.grid(False)
plot_confusion_matrix(rfc_grid_results['confusion_matrix'], classes=label, normalize=False, title='Confusion matrix', cmap=plt.cm.Greens)
plt.show()
```

```python
print_grid_search_attributes(rfc_grid_results['model'])
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import HistGradientBoostingClassifier

param_grid = {
    'max_depth': np.arange(5,8,1),
    'max_iter': np.arange(130,170,10)  # 'n_estimators' -> 'max_iter' for HistGradient
}

gbdt = HistGradientBoostingClassifier()
gbdt_random = RandomizedSearchCV(
    gbdt,
    param_distributions=param_grid,
    n_iter=5,
    n_jobs=-1,
    verbose=2,
    cv=3
)

gbdt_grid_results = perform_model(gbdt_random, X_train, y_train, X_test, y_test, class_labels=label)
```
