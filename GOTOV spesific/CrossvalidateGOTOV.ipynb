```python

# Load the CSV files
train_data = pd.read_csv('train_data.csv')
test_data = pd.read_csv('test_data.csv')

# Print to verify
print("Train shape:", train_data.shape)
print("Test shape:", test_data.shape)
```

```python
import pandas as pd

# Combine train and test data
combined_data = pd.concat([train_data, test_data], ignore_index=True)

# Function to create a split
def create_split(data, test_users):
    test_set = data[data['user_id'].isin(test_users)]
    train_set = data[~data['user_id'].isin(test_users)]
    return train_set, test_set

# Define user splits
split1_test_users = [1, 2, 3, 4]
split2_test_users = [10, 11, 12, 13, 14]
split3_test_users = [31, 32, 33, 34, 35, 36]

# Create the splits
train1, test1 = create_split(combined_data, split1_test_users)
train2, test2 = create_split(combined_data, split2_test_users)
train3, test3 = create_split(combined_data, split3_test_users)

# Verify
print("Split 1 - Train shape:", train1.shape, "Test shape:", test1.shape)
print("Split 2 - Train shape:", train2.shape, "Test shape:", test2.shape)
print("Split 3 - Train shape:", train3.shape, "Test shape:", test3.shape)
```

```python
label = train['label'].unique().tolist()
```

```python
from sklearn import linear_model
from sklearn.model_selection import GridSearchCV
from Set_up_for_performing import perform_model, print_grid_search_attributes

# Define Logistic Regression parameters
parameters = {'C': [0.01, 0.1, 1, 10, 20, 30], 'penalty': ['l2', 'l1']}

# List of train/test splits
splits = [(train1, test1), (train2, test2), (train3, test3)]

# Iterate over each split
for idx, (train_set, test_set) in enumerate(splits, 1):
    print(f"\n=== Evaluating Split {idx} ===")

    # Define X and y for the current split
    X_train, y_train = train_set.drop('label', axis=1), train_set['label']
    X_test, y_test = test_set.drop('label', axis=1), test_set['label']

    # Setup Grid Search
    log_reg = linear_model.LogisticRegression(solver='liblinear')
    log_reg_grid = GridSearchCV(log_reg, param_grid=parameters, cv=3, verbose=1, n_jobs=-1)

    # Perform training and evaluation
    log_reg_grid_results = perform_model(log_reg_grid, X_train, y_train, X_test, y_test, class_labels=label)

    # Print Grid Search results
    print_grid_search_attributes(log_reg_grid)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from Set_up_for_performing import perform_model, print_grid_search_attributes
import numpy as np

# Load datasets
train_data = pd.read_csv('train_data.csv')
test_data = pd.read_csv('test_data.csv')

# Combine datasets
combined_data = pd.concat([train_data, test_data], ignore_index=True)

# Function to create a split
def create_split(data, test_users):
    test_set = data[data['user_id'].isin(test_users)]
    train_set = data[~data['user_id'].isin(test_users)]
    return train_set, test_set

# Define user splits
splits = [
    [1, 2, 3, 4],
    [10, 11, 12, 13, 14],
    [31, 32, 33, 34, 35, 36]
]

# Random Forest parameters
params = {'n_estimators': np.arange(10, 201, 20), 'max_depth': np.arange(3, 15, 2)}

# Iterate over splits
for idx, test_users in enumerate(splits, 1):
    print(f"\n--- Evaluating Split {idx} ---")
    train_set, test_set = create_split(combined_data, test_users)

    # Split features and labels
    X_train, y_train = train_set.drop(['label', 'user_id'], axis=1), train_set['label']
    X_test, y_test = test_set.drop(['label', 'user_id'], axis=1), test_set['label']

    # Setup Grid Search
    rfc = RandomForestClassifier()
    rfc_grid = GridSearchCV(rfc, param_grid=params, cv=3, verbose=1, n_jobs=-1)

    # Perform training and evaluation
    rfc_grid_results = perform_model(rfc_grid, X_train, y_train, X_test, y_test, class_labels=y_train.unique())

    # Print grid search results
    print_grid_search_attributes(rfc_grid)
```
